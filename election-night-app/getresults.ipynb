```python
# importing the requests library 
import requests
import json
```

```python
# api-endpoint 
URL = "https://elections.maryland.gov/elections/results_data/GG18/Results.js"
page = requests.get(URL)
```

```python
data = page.json()
results = data['Results']
```

```python
print(results)
```

```python
results[0]
```

```python
hogan = []

for item in results:
    if item[0] == 12803 and item[2] == 2:
        hogan.append(item)
```

```python
print(hogan)
```

```python
jealous = []

for item in results:
    if item[0] == 12804 and item[2] == 2:
        jealous.append(item)
```

```python
print(jealous)
```

```python
with open('data.json', 'w') as outfile:
    json.dump(hogan, outfile)
```
